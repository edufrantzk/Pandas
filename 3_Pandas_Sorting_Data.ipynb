```python
import pandas as pd
```

```python
df = pd.read_csv('salarios.csv')
```

```python
df.head()
```

```python
#Ordenando pelo maior salário
df.sort_values(by=['Employee Annual Salary'], ascending=True, inplace=True)
df.drop(index=15111, inplace=True)
```

```python
df
```

```python
#Passando um objeto para tipo numérico(float)
df['Employee Annual Salary']= pd.to_numeric(df['Employee Annual Salary'],errors = 'coerce')
```

```python
df.info()
```

```python
#Ordenando pelo nome (de trás para frente)
df.sort_values(by=['Name'], ascending=False, inplace=True)
```

```python
df
```

```python
#Ordenando pelo índice
df.sort_index(inplace=True)
```
